```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout
```

```python
# Load the CSV files
df1 = pd.read_csv(r"C:\Users\abdel\PycharmProjects\Apple Detection\Dataset\csv files\train_annotations.csv")
df2 = pd.read_csv(r"C:\Users\abdel\PycharmProjects\Apple Detection\Dataset\csv files\test_annotations.csv")
df1.head()
```

```python
# Use a lambda function to create a new column 'category' based on the 'class' column
df1['category'] = df1['class'].apply(lambda x: 'Apple' if 'apple' in x.lower() else 'Damaged')
# Create a count plot
sns.set_style('darkgrid')
sns.countplot(x='category', data=df1, hue='class')
plt.show()
```

```python
# Define image folders
train_folder = r"C:\Users\abdel\PycharmProjects\Apple Detection\Dataset\train"
test_folder = r"C:\Users\abdel\PycharmProjects\Apple Detection\Dataset\test"
```

```python
# Select one image from the training set (assuming you want to display the first image)
sample_image = cv2.imread(r"C:\Users\abdel\PycharmProjects\Apple Detection\Dataset\train\apple--60-_jpg.rf.5ab650e7cc2a9518c40050e599d7c223.jpg")
# Display the image
plt.imshow(cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB))
plt.title("Normal Apple")
plt.show()

sample_image_2 = cv2.imread(r"C:\Users\abdel\PycharmProjects\Apple Detection\Dataset\train\damaged_apple--13-_jpg.rf.44fd493a468851c3745069eb9081af9c.jpg")
# Display the second image
plt.imshow(cv2.cvtColor(sample_image_2, cv2.COLOR_BGR2RGB))
plt.title("Damaged Apple")
plt.show()
```

```python
# Convert string labels to numeric values
label_encoder = LabelEncoder()
df1['class'] = label_encoder.fit_transform(df1['class'])
df2['class'] = label_encoder.transform(df2['class'])
df1['class'].head()
```

```python
def load_and_preprocess_images(df, image_folder, apply_augmentation=True):
    images = []
    labels = []

    for index, row in df.iterrows():
        image_path = os.path.join(image_folder, row['filename'])
        print(f"Loading image from: {image_path}")

        image = cv2.imread(image_path)
        if image is None:
            print(f"Error loading image: {image_path}")
            continue

        image = cv2.resize(image, (224, 224))
        image = image / 255.0

        if apply_augmentation:
            datagen = ImageDataGenerator(
                rotation_range=30,
                width_shift_range=0.1,
                height_shift_range=0.1,
                shear_range=0.2,
                zoom_range=0.2,
                horizontal_flip=True,
                vertical_flip=False
            )
            image = datagen.random_transform(image)

        images.append(image)
        labels.append(row['class'])

    return np.array(images), np.array(labels)
```

```python
# Split the dataset into training, validation, and test sets
train_df, val_df = train_test_split(df1, test_size=0.2, random_state=42)
```

```python
# Load and preprocess images for training and validation
train_images, train_labels = load_and_preprocess_images(train_df, train_folder, apply_augmentation=True)
```

```python
val_images, val_labels = load_and_preprocess_images(val_df, train_folder, apply_augmentation=False)  # No augmentation for validation
```

```python
# Define the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout((0.3)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout((0.2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()
```

```python
# Train the model
history = model.fit(train_images, train_labels, epochs=5, validation_data=(val_images, val_labels))
```

```python
# Plot training history
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Load and preprocess images for testing
test_images, test_labels = load_and_preprocess_images(df2, test_folder, apply_augmentation=False)
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(test_images, test_labels)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
import gradio as gr
import PIL.Image

def predict(image):
    # Check image type
    print(type(image))

    # Convert PIL Image to NumPy array if needed
    if isinstance(image, PIL.Image.Image):
        image = np.array(image)

    # Check image validity
    if image is None or image.size == 0:
        print("Error: Empty or invalid image.")
        return "Error: Empty or invalid image."

    # Preprocess the input image
    image = cv2.resize(image, (224, 224))
    image = image / 255.0
    image = np.expand_dims(image, axis=0)

    # Make predictions
    prediction = model.predict(image)

    # Convert the prediction to a class label
    predicted_class = "Apple" if prediction < 0.1 else "Damaged"

    print("The prediction:", predicted_class, "with confidence:", prediction)
    return predicted_class

# Create the Gradio interface
interface = gr.Interface(
    fn=predict,
    inputs=[gr.Image(type="pil", label="Upload Image")],
    outputs=gr.Label(num_top_classes=2, label="Prediction"),
    title="Prediction Apple",
    live=True
)

# Launch the interface
interface.launch()
```
